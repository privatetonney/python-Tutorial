# Python built in functions for arithmetic operations
These functions apply both to complex,boolean, integers and floats

abs(number/complex,): Returns the absolute value of a number.

min(iterable): Returns the smallest item in an iterable or the minimum of two or more arguments.

max(iterable): Returns the largest item in an iterable or the maximum of two or more arguments.

round(number, ndigits) Rounds a number to a given precision in decimal digits.(try positive and negative)

pow(base, exp):Returns base to the power of exp (equivalent to base**exp). 

sum(iterable):Returns the sum of all items in an iterable.

## 8. Lists


append(): Adds an element to the end of the list.

copy(): Returns a shallow copy of the list.

clear(): Removes all elements from the list.

count(): Returns the number of times a specified element appears in the list.

extend(): Adds elements from another list to the end of the current list.

index(): Returns the index of the first occurrence of a specified element.

insert(): Inserts an element at a specified position.

pop(): Removes and returns the element at the specified position (or the last element if no index is specified).

remove(): Removes the first occurrence of a specified element.

reverse(): Reverses the order of the elements in the list.

sort(): Sorts the list in ascending order (by default).

```python
#dir(list)
```

```python
my_list = [1, 2, 3, 'apple']
print(my_list)
my_list.append(10)
print('After append:', my_list)
print('Index 2:', my_list[2])
```

```python
#dir(set)
```

```python
# Creating a simple set
A = {1, 2, 3, 4, 5}
B={(1,2,"G")}
c={{1,2,3,4,5}}
print(A)
print(B)
```

```python
set1={1,2,3,4,5,6}

set2={8,9,10}


for ele in set2:
    set1.add(ele)
    
    
print(set1)    
set1.pop()
print(set1)
```

```python
#Sets Automatically Remove Duplicates

S = {2, 3,6,6,6,6,9,6, 3, 4}
print(S)
```

```python
# Adding Elements to a Set
A = {1, 2}
A.add(3)
print(A)
```

```python
#Removing Elements
A = {1, 2, 3}
A.remove(2)
print(A)

```

```python
# Union (A ∪ B)
A = {1, 2, 3}
B = {3, 4, 5}

print(A|B)
```

```python
#intersection (A ∩ B)
print(A & B)

```

```python
#Difference (A − B)
print(A - B)
```

```python
#Symmetric Difference (Elements not in both)
print(A ^ B)

```

```python
#Checking Membership
M = {1, 2, 3}
print(2 in M)
print(5 in M)
```

```python
#Subset Example
A = {1, 2, 3, 4}
B = {1, 2}

print(B.issubset(A))

```
