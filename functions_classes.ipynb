# Python Logical Operators: and & or
# 1️⃣ What Are Logical Operators?

Logical operators are used to combine conditional statements.

Python has three main logical operators:

and → True if both conditions are True.

or → True if at least one condition is True.

# 2️⃣ The and Operator

Returns True only if both operands are True.

Syntax:

```python
# condition1 and condition2 have to be true
#example
age = 25
income = 50000

if age > 18 and income > 30000:
    print("Eligible for loan")
else:
    print("Not eligible for loan")

```

# The or Operator

Returns True if at least one operand is True.

Syntax:
condition1 or condition2


```python
day = "Sunday"
holiday = True

if day == "Saturday" or holiday:
    print("You can relax today")
else:
    print("Go to work")
```

# Combining and & or

You can combine multiple logical operators.

Use parentheses to clarify precedence.

```python
x = 10
y = 20
z = 5

if (x > 5 and y > 15) or z > 10:
    print("Condition met")
else:
    print("Condition not met")
```

# Examples of use case scenarios 


```python
#Input validation
username = "admin"
password = "1234"

if username == "admin" and password == "1234":
    print("Login successful")
else:
    print("Login failed")

```

```python
# Range checks
number = 7
if number > 0 and number < 10:
    print("Number is between 1 and 9")
```

```python
#Multiple conditions
score = 85
grade = "A"

if score > 80 or grade == "A":
    print("Excellent!")
```

# Tips

and has higher precedence than or.

Use parentheses () to make conditions clear and readable.

Logical operators always return Boolean values (True or False).

# Arbitrary Keyword Arguments (**kwargs)

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=25, city="Paris")
```

# Recursion

A function that calls itself.

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))  # Output: 120
```

```python
#guessing game

def guess(num_of_guesses, correct_guess):
    count = 0
    while count < num_of_guesses:
        name = input("Enter your guess: ")
        if name == correct_guess:
            print("Congratulations! You won the game.")
            return  # End the function if correct
        else:
            print("Wrong guess. Try again.")
        count += 1

    print(f"Sorry, you lost the game. The correct answer was {correct_guess}.")
   
        
```

```python
guess(2,"cow")
```

## Python Classes Tutorial

# 1️⃣ Introduction to Classes

Definition: A class is a blueprint for creating objects.

Object: An instance of a class.

Classes allow encapsulation of data (attributes) and behavior (methods).

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

# Creating an object
person1 = Person("Alice", 25)
person1.greet()
```

# 2️⃣ Attributes and Methods

Attributes: Variables that belong to the object (self.name, self.age)

Methods: Functions inside a class that define behavior (greet())

```python
#Example

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

rect = Rectangle(4, 6)
print("Area:", rect.area())
print("Perimeter:", rect.perimeter())
```

# 3️⃣ Class vs Object

Class: Blueprint (Rectangle)

Object: Instance (rect)

You can create multiple objects from one class:

```python
rect1 = Rectangle(4, 6)
rect2 = Rectangle(10, 2)
```

# 4️⃣ Inheritance

A class can inherit from another class, reusing attributes and methods.

```python
class Square(Rectangle):
    def __init__(self, side):
        super().__init__(side, side)  # Reuse Rectangle constructor

sq = Square(5)
print("Square area:", sq.area())
```

# 5️⃣ Encapsulation (Private Attributes)

Use __ to make an attribute private:

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # private attribute

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

account = BankAccount(1000)
account.deposit(500)
print("Balance:", account.get_balance())
```

# 6️⃣ Polymorphism (Method Overriding)

A child class can override a method of the parent class:

```python
class Person:
    def greet(self):
        print("Hello!")

class Student(Person):
    def greet(self):
        print("Hi, I am a student!")

s = Student()
s.greet()
```
