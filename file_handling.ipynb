# Python Tutorial: Working With Files

Python provides built-in functions to create, read, write, and delete files.

 The open() Function
 
 open(filename, mode)
| Mode   | Meaning                                         |
| ------ | ----------------------------------------------- |
| `'r'`  | Read (default). File must exist.                |
| `'w'`  | Write. Creates new file OR overwrites existing. |
| `'a'`  | Append. Adds data to end of file.               |
| `'x'`  | Create a new file. Error if file exists.        |
| `'b'`  | Binary mode (images, audio).                    |
| `'t'`  | Text mode (default).                            |
| `'r+'` | Read and write.                                 |


# Reading Files

```python
# Reading Entire File

file = open("example.txt", "r")
content = file.read()
print(content)
file.close()
```

```python
# Reading Line by Line
file = open("data.txt", "r")
line = file.readline()
print(line)
file.close()
```

```python
# Using a loop

file = open("data.txt", "r")
for line in file:
    print(line.strip())  # removes newline
file.close()
```

```python
#Reading All Lines into a List

file = open("numbers.txt", "r")
lines = file.readlines()
print(lines)
file.close()
```

# Writing to Files

```python
# Writing (Overwrites File)
file = open("output.txt", "w")
file.write("Hello World!\n")
file.write("This is Python.")
file.close()
```

```python
# Appending (Adds to File)
file = open("output.txt", "a")
file.write("\nNew line added.")
file.close()

```

# Best Practice: Using with

Using with automatically closes the file.

```python
# Read File
with open("log.txt", "w") as f:
    f.write("Log started\n")

    
#Write File

with open("log.txt", "w") as f:
    f.write("Log started\n")

```
